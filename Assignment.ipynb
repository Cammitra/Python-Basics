```python
# Take user's name as input
name = input("Enter your name: ")

# Print greeting message
print("Hello,", name + "! Welcome to Python.")
```

```python
# Take input from the user
number = int(input("Enter a number: "))

# Check whether the number is even or odd
if number % 2 == 0:
    print(number, "is an Even number")
else:
    print(number, "is an Odd number")
```

```python
# Print numbers from 1 to 10

for number in range(1, 11):
    print(number)
```

```python
# Create a list of five numbers
numbers = [12, 45, 7, 89, 23]

# Find and print the maximum number
print("Maximum number is:", max(numbers))
```

```python
# Create a list with duplicate values
numbers = [1, 2, 3, 2, 4, 5, 1, 3]

# Remove duplicates using set
unique_numbers = list(set(numbers))

# Print the updated list
print("List after removing duplicates:", unique_numbers)
```

```python
def square(number):
    return number * number

# Example
print(square(5))
```

```python
# List of numbers
numbers = [2, 3, 4, 2, 5, 2]

# Number to count
x = 2

# Count how many times x appears
count = numbers.count(x)

# Print result
print("The number", x, "appears", count, "times.")
```
